```python
import re
import cv2
import numpy as np
import sqlite3
from paddleocr import PaddleOCR
from datetime import datetime, timedelta
from google.colab.patches import cv2_imshow

import sqlite3

# -----------------------------------------------------
# CREATE TABLE (SAFE OPEN + CLOSE)
# -----------------------------------------------------
with sqlite3.connect("grocery_ocr.db") as conn:
    cursor = conn.cursor()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS grocery_items (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        serial_no INTEGER,
        item TEXT,
        quantity TEXT,
        purchase_date DATE,
        expiry_date DATE,
        days_left INTEGER,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """)

    conn.commit()

print("Database Ready")

# Initialize OCR (RUN ONLY ONCE)
print("Initializing PaddleOCR...")
ocr = PaddleOCR(use_angle_cls=True, lang="en")
print("OCR Ready")
```

```python
print("Upload Grocery Bill Image")
uploaded = files.upload()

image_name = list(uploaded.keys())[0]

npimg = np.frombuffer(uploaded[image_name], np.uint8)
img = cv2.imdecode(npimg, cv2.IMREAD_COLOR)

def normalize_quantity(text):
    return re.sub(r"(\d+)([A-Za-z]+)", r"\1 \2", text)


def clean_item_name(item):
    item = item.lower()

    # Remove quantity patterns inside item
    item = re.sub(
        r"\b\d+\s*(kg|g|gm|liter|ltr|pack|packet|box|bottle|pcs|piece)\b",
        "",
        item,
        flags=re.IGNORECASE
    )

    item = re.sub(r"\b\d+\s*bo\b", "", item, flags=re.IGNORECASE)
    item = item.replace("-", " ")

    # Remove words like small, large
    item = re.sub(r"\bsmall\b", "", item)
    item = re.sub(r"\blarge\b", "", item)

    item = re.sub(r"\s+", " ", item).strip()

    # Normalize plural → singular basic fix
    if item.endswith("tomatoes"):
        item = "tomato"
    if item.endswith("caps"):
        item = "capsicum"
    if item.endswith("bestroot"):
        item="beetroot"

    return item


def is_quantity(text):
    text = normalize_quantity(text)

    pattern = r"^\d+\s*(kg|kgs|gm|g|liter|ltr|pack|packet|box|bottle|pcs|piece)$"

    return bool(re.match(pattern, text.strip(), re.IGNORECASE))

def calculate_expiry(item_name):

    today = datetime.today().date()

    row = shelf_df[
        shelf_df["Ingredient"] == item_name.lower()
    ]

    if not row.empty:
        fridge_days = int(row["Refrigerator_Shelf_Life_Days"].values[0])
        pantry_days = int(row["Pantry_Shelf_Life_Days"].values[0])

        # 70% fridge + 30% pantry
        avg_days = round((0.7 * fridge_days) + (0.3 * pantry_days))

        expiry_date = today + timedelta(days=avg_days)

        return today, expiry_date, avg_days

    # Default fallback
    default_days = 5
    expiry_date = today + timedelta(days=default_days)
    return today, expiry_date, default_days



def process_grocery_list(words):

    grocery_items = []

    skip_words = [
        "subtotal", "delivery", "charge", "order",
        "paid", "total", "summary", "price","DMart","Vegetables & Greens","Meat& Seafood","Diary & Refrigerated","Grains,Pluses & Flours","Oils & Liquids","Spices& Seasonings",
        "track", "www", "jm","Grand","Total","Items","Idli & Dosa Essentials","Vegetables & Herbs","Spices & Seasonings","Other Ingredients","For Snacks & Chutney"
    ]

    i = 0
    while i < len(words) - 1:

        item = words[i].strip()
        next_word = words[i + 1].strip()

        # Skip unwanted header/footer words
        if any(x in item.lower() for x in skip_words):
            i += 1
            continue

        # Skip if item is purely numeric (like 30)
        if item.isdigit():
            i += 1
            continue

        #  Skip if item contains mostly digits (e.g. 1,580)
        if re.match(r"^\d+[,\d]*$", item):
            i += 1
            continue

        if is_quantity(next_word):

            clean_item = clean_item_name(item)
            clean_qty = normalize_quantity(next_word)

            #  Extra safety: skip empty or numeric after cleaning
            if clean_item == "" or clean_item.isdigit():
                i += 2
                continue

            grocery_items.append([clean_item, clean_qty])
            i += 2
        else:
            i += 1

    return grocery_items


```

```python
result = ocr.ocr(img)

extracted_text = []
for line in result:
    for word in line:
        text = word[1][0].strip()
        if text:
            extracted_text.append(text)

print("Extracted OCR Words:")
print(extracted_text)

grocery_data = process_grocery_list(extracted_text)

if not grocery_data:
    print("No grocery items detected!")
else:
    import pandas as pd

    df = pd.DataFrame(grocery_data, columns=["Item", "Quantity"])
    df = df.drop_duplicates()
    df.insert(0, "S.No", range(1, len(df) + 1))

    display(df)
# -----------------------------------------------------
# SAVE TO DATABASE (SAFE CONNECTION)
# -----------------------------------------------------
with sqlite3.connect("grocery_ocr.db") as conn:
    cursor = conn.cursor()

    for _, row in df.iterrows():
        purchase_date, expiry_date, days_left = calculate_expiry(row["Item"])

        cursor.execute("""
            INSERT INTO grocery_items
            (serial_no, item, quantity, purchase_date, expiry_date, days_left)
            VALUES (?, ?, ?, ?, ?, ?)
        """, (
            row["S.No"],
            row["Item"],
            row["Quantity"],
            purchase_date,
            expiry_date,
            days_left
        ))

    conn.commit()

print("Data Saved Successfully!")
```

```python
# =====================================================
# SMART INVENTORY INTELLIGENCE LAYER (FINAL CORRECT)
# =====================================================

import pandas as pd
import sqlite3
from datetime import datetime

# -----------------------------------------------------
# 1️⃣ Connect to Database
# -----------------------------------------------------
# -----------------------------------------------------
# CONNECT & LOAD DATA SAFELY
# -----------------------------------------------------
with sqlite3.connect("grocery_ocr.db") as conn:
    df = pd.read_sql("SELECT * FROM grocery_items", conn)

if df.empty:
    print("No inventory data found.")
    raise SystemExit()
# -----------------------------------------------------
# 2️⃣ Clean & Prepare Data
# -----------------------------------------------------

# Ensure expiry_date is datetime
df["expiry_date"] = pd.to_datetime(df["expiry_date"], errors="coerce")

# Extract numeric quantity safely
df["qty_value"] = (
    df["quantity"]
    .str.extract(r"(\d+\.?\d*)")[0]
    .astype(float)
)

# Drop rows with invalid expiry
df = df.dropna(subset=["expiry_date"])

# -----------------------------------------------------
# 3️⃣ Aggregate Inventory (Merge Duplicates)
# -----------------------------------------------------

df_grouped = (
    df.groupby("item", as_index=False)
      .agg({
          "qty_value": "sum",          # total quantity
          "expiry_date": "min"         # earliest expiry
      })
)

# -----------------------------------------------------
# 4️⃣ Recalculate Days Left
# -----------------------------------------------------

today = pd.Timestamp(datetime.today().date())

df_grouped["expiry_date"] = pd.to_datetime(
    df_grouped["expiry_date"], errors="coerce"
)

df_grouped["days_left"] = (
    df_grouped["expiry_date"] - today
).dt.days

# -----------------------------------------------------
# 5️⃣ Expiry Status Logic
# -----------------------------------------------------

def get_status(days):
    if pd.isna(days):
        return "Unknown"
    elif days < 0:
        return "Expired"
    elif days <= 3:
        return "Urgent"
    elif days <= 7:
        return "Use Soon"
    else:
        return "Fresh"

df_grouped["status"] = df_grouped["days_left"].apply(get_status)

# -----------------------------------------------------
# 6️⃣ Smart Priority Score
# -----------------------------------------------------

def priority_score(days):
    if pd.isna(days):
        return 0
    if days < 0:
        return 1000  # highest urgency
    return round(1 / (days + 1), 4)

df_grouped["priority_score"] = df_grouped["days_left"].apply(priority_score)

# -----------------------------------------------------
# 7️⃣ Smart Sorting (Real Priority Order)
# -----------------------------------------------------

status_order = {
    "Expired": 0,
    "Urgent": 1,
    "Use Soon": 2,
    "Fresh": 3,
    "Unknown": 4
}

df_grouped["status_rank"] = df_grouped["status"].map(status_order)

df_grouped = df_grouped.sort_values(
    by=["status_rank", "days_left"]
).reset_index(drop=True)

# -----------------------------------------------------
# 8️⃣ Final Display
# -----------------------------------------------------

print("\n SMART INVENTORY (MERGED VIEW):\n")
display(df_grouped[[
    "item",
    "qty_value",
    "expiry_date",
    "days_left",
    "status",
    "priority_score"
]])

print("\n TOP 5 ITEMS TO USE FIRST:\n")
display(df_grouped.head(5)[[
    "item",
    "qty_value",
    "days_left",
    "status"
]])

```

```python
# CLEAN & NORMALIZE CUISINE

def normalize_cuisine(text):
    text = str(text).lower().strip()

    cuisine_map = {
        "andhra": "Andhra",
        "hyderabadi": "Hyderabadi",
        "south indian": "South Indian",
        "north indian": "North Indian",
        "bengali": "Bengali",
        "gujarati": "Gujarati",
        "indian": "Indian",
        "karnataka": "Karnataka",
        "kashmiri": "Kashmiri",
        "kerala": "Kerala",
        "lucknowi": "Lucknowi",
        "maharashtrian": "Maharashtrian",
        "nagaland": "Nagaland",
        "uttar pradesh": "Uttar Pradesh",
        "udupi": "Udupi",
        "tamil nadu": "Tamil Nadu",
        "rajasthani": "Rajasthani",
        "punjabi": "Punjabi",
        "himachal": "Himachal",
        "jharkhand": "Jharkhand",
        "oriya": "Odia"
    }

    for key in cuisine_map:
        if key in text:
            return cuisine_map[key]

    return "Other"


# Apply cleaning
df_small["Cuisine_Clean"] = df_small["Cuisine"].apply(normalize_cuisine)

print("Cuisine_Clean column created successfully!")
df_small[["Cuisine", "Cuisine_Clean"]].head()
```

```python
# Normalize Course into Standard Meal Categories

def normalize_meal(text):
    text = str(text).lower().strip()

    # Time-based meals
    if "breakfast" in text:
        return "Breakfast"
    elif "lunch" in text:
        return "Lunch"
    elif "dinner" in text:
        return "Dinner"

    # Course-based meals
    elif any(word in text for word in ["main course", "curry", "biryani", "rice", "roti", "paratha"]):
        return "Main Course"

    elif any(word in text for word in ["side dish", "accompaniment", "salad", "raita"]):
        return "Side Dish"

    elif any(word in text for word in ["dessert", "sweet", "cake", "halwa", "kheer"]):
        return "Dessert"

    elif any(word in text for word in ["snack", "starter", "appetizer", "chaat", "pakora"]):
        return "Snack"

    else:
        return "Other"


# Apply function
df_small["MealType"] = df_small["Course"].apply(normalize_meal)

print("MealType column created successfully!")
```

```python
import re
import pandas as pd
import unicodedata

REMOVE_WORDS = {
    "cup","cups","tablespoon","tablespoons","tbsp","teaspoon","teaspoons","tsp",
    "gram","grams","kg","g","ml","liter","litre",
    "inch","inches","pinch","clove","cloves","slice","sprig",
    "chopped","sliced","diced","finely","roughly","picked","crushed","roasted",
    "small","large","medium","optional","fresh","whole",
    "taste","required","need","needed",
    "powder","paste","ground","yellow",
    "handful","choice","like","white","sprig",
    "and","with","for","to","as"
}

# Common OCR fixes dictionary
OCR_FIXES = {
    "cor iander": "coriander",
    "peppercor ns": "peppercorns",
    "ch illi": "chilli",
    "gi nger": "ginger"
}

def clean_ingredient(item):

    item = str(item).strip().lower()
    item = unicodedata.normalize("NFKC", item)

    # Remove bracket content
    item = re.sub(r"\([^)]*\)", "", item)

    # Remove numbers & fractions
    item = re.sub(r"\d+\/?\d*", "", item)

    # Fix joined "orhoney"
    item = re.sub(r"or(?=[a-z])", "or ", item)

    # Keep only first alternative
    item = item.split(" or ")[0]

    # Remove text after hyphen
    item = item.split("-")[0]

    # Remove special characters (keep letters & spaces)
    item = re.sub(r"[^a-z\s]", "", item)

    # Fix known OCR issues
    for wrong, correct in OCR_FIXES.items():
        item = item.replace(wrong, correct)

    # Remove unwanted words
    words = [w for w in item.split() if w not in REMOVE_WORDS]

    item = " ".join(words)

    # Normalize common ingredient patterns

    item = item.replace("pepper corn", "peppercorn")
    item = item.replace("mint leaf", "mint leaves")
    item = item.replace("curry leaf", "curry leaves")

    # Remove extra spaces
    item = re.sub(r"\s+", " ", item).strip()

    return item
```

```python
df_clean["Ingredients"] = (
    df_clean["Ingredients"]
    .fillna("")
    .astype(str)
)

cleaned_ingredients = []

for row in df_clean["Ingredients"]:

    items = row.split(",")
    temp = []

    for item in items:
        cleaned = clean_ingredient(item)
        if cleaned:
            temp.append(cleaned)

    # Remove duplicates while keeping order
    temp = list(dict.fromkeys(temp))

    cleaned_ingredients.append(temp)

df_clean["Cleaned_Ingredients"] = cleaned_ingredients
```

```python
# ==========================================================
# IMPORTS
# ==========================================================

import numpy as np
import sqlite3
import re
import pandas as pd
import spacy
import matplotlib.pyplot as plt

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from rapidfuzz import process, fuzz

import ipywidgets as widgets
from IPython.display import display, clear_output

# PDF
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, ListFlowable, ListItem
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.units import inch
from reportlab.lib.pagesizes import letter

from google.colab import files


# ==========================================================
# MODEL LOADING
# ==========================================================

print("🔄 [STEP 1/5] Loading SentenceTransformer model...")
bi_encoder = SentenceTransformer("all-MiniLM-L6-v2")
print("✅ SentenceTransformer loaded successfully.")

print("🔄 [STEP 2/5] Loading SpaCy model...")
nlp = spacy.load("en_core_web_sm")
print("✅ SpaCy model loaded successfully.\n")

IGNORE_INGREDIENTS = {"water", "salt", "oil", "sunflower oil"}
REMOVE_FROM_MISSED = {"red", "a", "break wheat"}


# ==========================================================
# NORMALIZATION
# ==========================================================

def normalize_ingredient(text):
    text = str(text).lower().strip()
    text = re.sub(r'\s+', ' ', text)

    replacements = {
        "cor n": "corn",
        "podsseeds": "pods",
        "cardamompod": "cardamom pod",
        "chillies": "chilli",
        "leaves": "leaves",
        "kashmiri red chilli":"red chilli powder"
    }

    for k, v in replacements.items():
        text = text.replace(k, v)

    text = re.sub(r'[^a-zA-Z\s]', '', text)

    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc if not token.is_stop]

    return " ".join(lemmas)


# ==========================================================
# LOAD INVENTORY
# ==========================================================

def load_inventory(db_path="grocery_ocr.db"):

    print("🔄 [STEP 3/5] Loading inventory database...")

    conn = sqlite3.connect(db_path)
    df_inv = pd.read_sql_query("SELECT item, days_left FROM grocery_items", conn)
    conn.close()

    if df_inv.empty:
        print("⚠️ Inventory is empty.\n")
        return [], {}, ""

    df_inv["item"] = df_inv["item"].apply(normalize_ingredient)
    df_inv = df_inv.groupby("item", as_index=False)["days_left"].min()

    inventory_dict = dict(zip(df_inv["item"], df_inv["days_left"]))
    inventory_items = [i for i in inventory_dict if i not in IGNORE_INGREDIENTS]

    print(f"✅ Loaded {len(inventory_items)} inventory items.\n")

    return inventory_items, inventory_dict, " ".join(inventory_items)


# ==========================================================
# FUZZY MATCHING
# ==========================================================

def find_inventory_match(ingredient, inventory_keys, score_cutoff=85):

    if ingredient in inventory_keys:
        return ingredient

    match = process.extractOne(
        ingredient,
        inventory_keys,
        scorer=fuzz.token_sort_ratio,
        score_cutoff=score_cutoff
    )

    return match[0] if match else None


# ==========================================================
# RECOMMENDER CLASS
# ==========================================================

class RecipeRecommender:

    def __init__(self, df_clean, inventory_db="grocery_ocr.db"):

        print("🔄 [STEP 4/5] Preparing recipe dataset...")

        self.df_clean = df_clean.copy().drop_duplicates(subset=["RecipeName"]).reset_index(drop=True)

        self.df_clean["Cleaned_Ingredients"] = self.df_clean["Cleaned_Ingredients"].apply(
            lambda x: [normalize_ingredient(i) for i in x] if isinstance(x, list) else []
        )

        self.df_clean["recipe_text"] = self.df_clean["Cleaned_Ingredients"].apply(
            lambda x: " ".join([i for i in x if i not in IGNORE_INGREDIENTS])
        )

        self.df_clean["Cuisine_Clean"] = self.df_clean["Cuisine_Clean"].fillna("Unknown")
        self.df_clean["Diet"] = self.df_clean["Diet"].fillna("Unknown")
        self.df_clean["MealType"] = self.df_clean["MealType"].fillna("Unknown")
        self.df_clean["TranslatedInstructions"] = self.df_clean["TranslatedInstructions"].fillna("No instructions available.")

        print(f"✅ Dataset ready with {len(self.df_clean)} recipes.\n")

        self.inventory_items, self.inventory_dict, self.inventory_text = load_inventory(inventory_db)

        print("🔄 [STEP 5/5] Encoding inventory & recipes (CPU mode)...")

        self.inventory_emb = bi_encoder.encode(self.inventory_text, convert_to_numpy=True)

        self.recipe_embeddings = bi_encoder.encode(
            self.df_clean["recipe_text"].tolist(),
            convert_to_numpy=True,
            batch_size=128,
            show_progress_bar=True
        )

        print("✅ Encoding complete.")
        print("🎉 System Ready!\n")

        self.filtered_df = self.df_clean.copy()
        self.index = 0
        self.current_row = None

        self.setup_widgets()


    # ==========================================================
    # STRICT FILTER
    # ==========================================================
    def strict_filter(self, df):

        c = self.cuisine_dd.value
        d = self.diet_dd.value
        m = self.meal_dd.value

        level1 = df.copy()

        if c != "Any":
            level1 = level1[level1["Cuisine_Clean"].str.lower() == c.lower()]
        if d != "Any":
            level1 = level1[level1["Diet"].str.lower() == d.lower()]
        if m != "Any":
            level1 = level1[level1["MealType"].str.lower() == m.lower()]

        if not level1.empty:
            return level1, "Matched Meal + Diet + Cuisine"

        if d != "Any" and m != "Any":
            level2 = df[
                (df["MealType"].str.lower() == m.lower()) &
                (df["Diet"].str.lower() == d.lower())
            ]
            if not level2.empty:
                return level2, "Matched Meal + Diet"

        if m != "Any":
            level3 = df[df["MealType"].str.lower() == m.lower()]
            if not level3.empty:
                return level3, "Matched Meal Type only"

        if d != "Any":
            level4 = df[df["Diet"].str.lower() == d.lower()]
            if not level4.empty:
                return level4, "Matched Diet only"

        return df, "No exact match found. Showing best overall."


    # ==========================================================
    # SCORING
    # ==========================================================
    def compute_scores(self, df):

        indices = df.index.tolist()

        sims = cosine_similarity(
            self.recipe_embeddings[indices],
            [self.inventory_emb]
        ).flatten()

        df = df.copy()
        df["Semantic_Score"] = sims

        matched_list, missed_list = [], []
        coverage_list, expiry_list = [], []

        for idx in indices:

            ingredients = self.df_clean.loc[idx, "Cleaned_Ingredients"]
            filtered = [i for i in ingredients if i not in IGNORE_INGREDIENTS]

            matched = []
            expiry_scores = []

            for ing in filtered:
                match = find_inventory_match(ing, self.inventory_dict.keys())
                if match:
                    matched.append(match)
                    days = self.inventory_dict.get(match, 30)
                    expiry_scores.append(max(0, (30 - days)/30))

            matched = list(set(matched))
            missed = list(set(filtered) - set(matched))
            missed = [m for m in missed if m not in REMOVE_FROM_MISSED]

            coverage = len(matched) / len(filtered) if filtered else 0
            expiry = np.mean(expiry_scores) if expiry_scores else 0

            matched_list.append(matched)
            missed_list.append(missed)
            coverage_list.append(round(coverage,3))
            expiry_list.append(round(expiry,3))

        df["Matched_Ingredients"] = matched_list
        df["Missed_Ingredients"] = missed_list
        df["Coverage_Score"] = coverage_list
        df["Expiry_Score"] = expiry_list

        df["Final_Score"] = (
            0.4 * df["Semantic_Score"] +
            0.4 * df["Coverage_Score"] +
            0.2 * df["Expiry_Score"]
        )

        return df.sort_values("Final_Score", ascending=False).reset_index(drop=True)


    # ==========================================================
    # UI
    # ==========================================================
    def setup_widgets(self):

        cuisines = ["Any"] + sorted(self.df_clean["Cuisine_Clean"].unique())
        diets = ["Any"] + sorted(self.df_clean["Diet"].unique())
        meals = ["Any"] + sorted(self.df_clean["MealType"].unique())

        self.cuisine_dd = widgets.Dropdown(options=cuisines, description="Cuisine:")
        self.diet_dd = widgets.Dropdown(options=diets, description="Diet:")
        self.meal_dd = widgets.Dropdown(options=meals, description="Meal:")

        self.recommend_btn = widgets.Button(description="Recommend", button_style="primary")
        self.next_btn = widgets.Button(description="Next", button_style="success")
        self.reset_btn = widgets.Button(description="Reset Preferences", button_style="warning")
        self.download_recipe_btn = widgets.Button(description="📥 Download Recipe PDF", button_style="info")
        self.download_grocery_btn = widgets.Button(description="🛒 Download Grocery PDF", button_style="info")

        self.out = widgets.Output()

        self.recommend_btn.on_click(self.recommend_recipes)
        self.next_btn.on_click(self.show_next)
        self.reset_btn.on_click(self.reset_preferences)
        self.download_recipe_btn.on_click(self.download_recipe_pdf)
        self.download_grocery_btn.on_click(self.download_grocery_pdf)

        display(widgets.VBox([
            widgets.HBox([self.cuisine_dd, self.diet_dd, self.meal_dd]),
            widgets.HBox([
                self.recommend_btn,
                self.next_btn,
                self.reset_btn,
                self.download_recipe_btn,
                self.download_grocery_btn
            ]),
            self.out
        ]))


    # ==========================================================
    # ACTIONS
    # ==========================================================
    def recommend_recipes(self, b):
        df, self.explanation = self.strict_filter(self.df_clean)
        self.filtered_df = self.compute_scores(df)
        self.index = 0
        self.show_next(None)


    def show_next(self, b):

        if self.filtered_df.empty:
            with self.out:
                clear_output()
                print("No recipes found.")
            return

        row = self.filtered_df.iloc[self.index % len(self.filtered_df)]
        self.index += 1
        self.current_row = row

        with self.out:
            clear_output()

            print("Recipe:", row["RecipeName"])
            print("Cuisine:", row["Cuisine_Clean"])
            print("Meal Type:", row["MealType"])
            print("Diet:", row["Diet"])
            print("Semantic Score:", round(row["Semantic_Score"], 3))
            print("Coverage Score:", row["Coverage_Score"])
            print("Expiry Score:", row["Expiry_Score"])
            print("Final Score:", round(row["Final_Score"], 3))

            print("\nMatched Ingredients:", ", ".join(row["Matched_Ingredients"]))
            if row["Missed_Ingredients"]:
                print("Missed Ingredients:", ", ".join(row["Missed_Ingredients"]))

            print("\nWhy this recipe?", self.explanation)
            print("\nInstructions:\n", row["TranslatedInstructions"])

            print("\n[INFO] Recommendation ready!")


    def reset_preferences(self, b):
        self.cuisine_dd.value = "Any"
        self.diet_dd.value = "Any"
        self.meal_dd.value = "Any"
        with self.out:
            clear_output()
            print("✅ Preferences reset successfully!")


    # ==========================================================
    # PDF DOWNLOADS
    # ==========================================================
    def download_recipe_pdf(self, b):

        if self.current_row is None:
            return

        row = self.current_row
        pdf_path = "Recommended_Recipe.pdf"

        doc = SimpleDocTemplate(pdf_path, pagesize=letter)
        elements = []
        styles = getSampleStyleSheet()

        elements.append(Paragraph(f"Recipe: {row['RecipeName']}", styles["Heading1"]))
        elements.append(Spacer(1, 0.3 * inch))

        elements.append(Paragraph("Matched Ingredients:", styles["Heading2"]))
        elements.append(ListFlowable(
            [ListItem(Paragraph(i, styles["Normal"])) for i in row["Matched_Ingredients"]],
            bulletType='bullet'
        ))

        elements.append(Spacer(1, 0.3 * inch))

        if row["Missed_Ingredients"]:
            elements.append(Paragraph("Grocery Items Needed:", styles["Heading2"]))
            elements.append(ListFlowable(
                [ListItem(Paragraph(i, styles["Normal"])) for i in row["Missed_Ingredients"]],
                bulletType='bullet'
            ))

        elements.append(Spacer(1, 0.3 * inch))
        elements.append(Paragraph("Instructions:", styles["Heading2"]))
        elements.append(Paragraph(row["TranslatedInstructions"], styles["Normal"]))

        doc.build(elements)
        files.download(pdf_path)


    def download_grocery_pdf(self, b):

        if self.current_row is None:
            return

        row = self.current_row
        pdf_path = "Grocery_List.pdf"

        doc = SimpleDocTemplate(pdf_path, pagesize=letter)
        elements = []
        styles = getSampleStyleSheet()

        elements.append(Paragraph("Grocery List", styles["Heading1"]))
        elements.append(Spacer(1, 0.3 * inch))

        if row["Missed_Ingredients"]:
            elements.append(ListFlowable(
                [ListItem(Paragraph(i, styles["Normal"])) for i in row["Missed_Ingredients"]],
                bulletType='bullet'
            ))
        else:
            elements.append(Paragraph("No additional items needed.", styles["Normal"]))

        doc.build(elements)
        files.download(pdf_path)


```

```python
def evaluate_recommendations(recommender, top_k=10):
    """
    Evaluates the recipe recommendation accuracy based on ingredient coverage.
    """
    df = recommender.filtered_df.copy().head(top_k)

    avg_coverage = df["Coverage_Score"].mean()
    avg_expiry = df["Expiry_Score"].mean()
    avg_final_score = df["Final_Score"].mean()
    avg_missed_ingredients = df["Missed_Ingredients"].apply(len).mean()
    avg_matched_ingredients = df["Matched_Ingredients"].apply(len).mean()

    print("========== Recommendation Evaluation Metrics ==========")
    print(f"Average Coverage Score: {avg_coverage:.3f}")
    print(f"Average Expiry Score: {avg_expiry:.3f}")
    print(f"Average Final Score: {avg_final_score:.3f}")
    print(f"Average Matched Ingredients: {avg_matched_ingredients:.2f}")
    print(f"Average Missed Ingredients: {avg_missed_ingredients:.2f}")
    print("======================================================")

    return {
        "avg_coverage": avg_coverage,
        "avg_expiry": avg_expiry,
        "avg_final_score": avg_final_score,
        "avg_matched_ingredients": avg_matched_ingredients,
        "avg_missed_ingredients": avg_missed_ingredients
    }
```

```python
# -------------------------
# Full Day Meal Planner with Leave-One-Out Dynamic Grocery List
# -------------------------
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

class FullDayMealPlannerDynamic:
    def __init__(self, recommender):
        """
        recommender: instance of RecipeRecommender with:
                     - df_clean: preprocessed recipes DataFrame
                     - compute_scores(df): method returning df with 'Final_Score', 'Expiry_Score', 'Missed_Ingredients'
        """
        self.recommender = recommender
        self.meal_types = ["Breakfast", "Lunch", "Dinner"]
        self.meal_indices = {meal: 0 for meal in self.meal_types}
        self.meal_dfs = {meal: pd.DataFrame() for meal in self.meal_types}
        self.grocery_list = set()

        # Buttons for each meal
        self.refresh_buttons = {
            meal: widgets.Button(description=f"Refresh {meal}", button_style="info")
            for meal in self.meal_types
        }
        for meal, btn in self.refresh_buttons.items():
            btn.on_click(lambda b, m=meal: self.show_meal(m, leave_one_out=True))

        # Output areas
        self.out = widgets.Output()
        self.grocery_out = widgets.Output()

        # Display UI
        display(widgets.VBox([
            widgets.HBox(list(self.refresh_buttons.values())),
            self.out,
            widgets.Label("🛒 Daily Grocery List:"),
            self.grocery_out
        ]))

        # Generate initial meal plan
        self.generate_meal_plan()

    # -------------------------
    # Generate meal plan with expiry-aware priority
    # -------------------------
    def generate_meal_plan(self):
        self.grocery_list = set()
        for meal in self.meal_types:
            df_meal = self.recommender.df_clean[
                self.recommender.df_clean["MealType"].str.lower() == meal.lower()
            ]
            df_meal_filtered = self.recommender.compute_scores(df_meal)

            # Sort by expiry and final score
            df_meal_filtered = df_meal_filtered.sort_values(
                by=["Expiry_Score", "Final_Score"], ascending=[False, False]
            ).reset_index(drop=True)

            self.meal_dfs[meal] = df_meal_filtered
            self.meal_indices[meal] = 0
            self.show_meal(meal, initial=True)

    # -------------------------
    # Show meal (with optional Leave-One-Out refresh)
    # -------------------------
    def show_meal(self, meal, initial=False, leave_one_out=False):
        df = self.meal_dfs[meal]
        if df.empty:
            with self.out:
                clear_output(wait=True)
                print(f"No recipes available for {meal}")
            self.show_grocery_list()
            return

        # Determine which recipe to show
        idx = self.meal_indices[meal]
        row = df.iloc[idx % len(df)]
        if leave_one_out and not initial:
            # Cycle to the next recipe
            self.meal_indices[meal] = (idx + 1) % len(df)
            idx = self.meal_indices[meal]
            row = df.iloc[idx % len(df)]

        # Update grocery list with current meal's missing ingredients
        missed = row.get("Missed_Ingredients", [])
        if isinstance(missed, list):
            self.grocery_list.update(missed)

        # Display meal plan
        with self.out:
            clear_output(wait=True)
            print("🍽 FULL DAY MEAL PLAN")
            print("="*50)
            for m in self.meal_types:
                m_idx = self.meal_indices[m] if m != meal else idx
                m_df = self.meal_dfs[m]
                if m_df.empty:
                    print(f"{m}: No recipe available")
                    continue
                m_row = m_df.iloc[m_idx % len(m_df)]
                print(f"\n{m.upper()}")
                print(f"Recipe: {m_row['RecipeName']}")
                print(f"Cuisine: {m_row['Cuisine_Clean']}")
                print(f"Meal Type: {m_row['MealType']}")
                instructions = m_row.get("TranslatedInstructions", None)
                if instructions and pd.notna(instructions):
                    print("----------------------------------------------------")
                    print("📖 Instructions:")
                    print(instructions)
                print("="*50)

        # Show consolidated grocery list
        self.show_grocery_list()

    # -------------------------
    # Show consolidated grocery list
    # -------------------------
    def show_grocery_list(self):
        with self.grocery_out:
            clear_output(wait=True)
            if self.grocery_list:
                unique_grocery_list = sorted(set(self.grocery_list))
                print("🛒 DAILY GROCERY LIST (Unique Items):")
                print("="*40)
                for item in unique_grocery_list:
                    print(f"- {item}")
            else:
                print("No missing ingredients! 🛒")
```

```python
import numpy as np
import sqlite3
import re
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# -------------------------
# 1️⃣ Load Inventory
# -------------------------
def load_inventory(db_path="grocery_ocr.db"):
    print("[INFO] Loading inventory from database...")
    conn = sqlite3.connect(db_path)
    df_inv = pd.read_sql_query("SELECT item, days_left FROM grocery_items", conn)
    conn.close()

    if df_inv.empty:
        print("[WARNING] Inventory is empty!")
        return [], {}, "", pd.DataFrame(columns=["item","days_left"])

    df_inv["item"] = df_inv["item"].str.lower().str.strip()
    df_inv = df_inv.groupby("item", as_index=False)["days_left"].min()
    inventory_dict = dict(zip(df_inv["item"], df_inv["days_left"]))
    inventory_items = list(inventory_dict.keys())
    inventory_text = " ".join(inventory_items)

    print(f"[INFO] Inventory loaded: {len(inventory_items)} items found.")
    return inventory_items, inventory_dict, inventory_text, df_inv

# -------------------------
# 2️⃣ Text Cleaning
# -------------------------
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"\s+", " ", text)
    return text.strip()

# -------------------------
# 3️⃣ Preprocess Recipes
# -------------------------
def preprocess_recipes(df):
    print("[INFO] Preprocessing recipes...")
    df = df.drop_duplicates(subset=["RecipeName"]).reset_index(drop=True)
    df["Cleaned_Ingredients"] = df["Cleaned_Ingredients"].apply(
        lambda x: [clean_text(i) for i in x] if isinstance(x, list) else []
    )
    df["ingredient_text"] = df["Cleaned_Ingredients"].apply(lambda x: " ".join(x))
    df["Cuisine_Clean"] = df["Cuisine_Clean"].fillna("Unknown")
    df["Diet"] = df["Diet"].fillna("Unknown")
    df["MealType"] = df["MealType"].fillna("Unknown")
    print(f"[INFO] Recipes preprocessed: {len(df)} recipes.")
    return df

# -------------------------
# 4️⃣ Interactive TF-IDF Recipe Recommender
# -------------------------
IGNORE_INGREDIENTS = set(["water", "sunflower oil", "salt", "oil"])

class InteractiveRecipeRecommenderTFIDF:
    def __init__(self, df_clean, inventory_db="grocery_ocr.db"):
        self.df_clean = preprocess_recipes(df_clean)

        # Load inventory
        self.inventory_items, self.inventory_dict, self.inventory_text, _ = load_inventory(inventory_db)
        self.inventory_items_filtered = [item for item in self.inventory_items if item not in IGNORE_INGREDIENTS]

        # Combine inventory text
        self.inventory_text_filtered = " ".join(self.inventory_items_filtered)

        # TF-IDF vectorizer
        self.vectorizer = TfidfVectorizer()
        print("[INFO] Fitting TF-IDF vectorizer...")
        corpus = self.df_clean["ingredient_text"].tolist() + [self.inventory_text_filtered]
        self.vectorizer.fit(corpus)
        self.inventory_vec = self.vectorizer.transform([self.inventory_text_filtered])
        print("[INFO] TF-IDF vectorizer ready.")

        self.filtered_df = self.df_clean.copy()
        self.index = 0
        self.setup_widgets()

    # -------------------------
    # Multi-level filtering
    # -------------------------
    def strict_filter(self, df):
        c, d, m = self.cuisine_dd.value, self.diet_dd.value, self.meal_dd.value
        level1 = df.copy()
        if c != "Any":
            level1 = level1[level1["Cuisine_Clean"].str.lower() == c.lower()]
        if d != "Any":
            level1 = level1[level1["Diet"].str.lower() == d.lower()]
        if m != "Any":
            level1 = level1[level1["MealType"].str.lower() == m.lower()]
        if not level1.empty:
            return level1, "Matched Meal + Diet + Cuisine"

        # Level 2
        if d != "Any" and m != "Any":
            level2 = df[(df["MealType"].str.lower() == m.lower()) & (df["Diet"].str.lower() == d.lower())]
            if not level2.empty:
                return level2, "Matched Meal + Diet"

        # Level 3
        if m != "Any":
            level3 = df[df["MealType"].str.lower() == m.lower()]
            if not level3.empty:
                return level3, "Matched Meal Type only"

        # Level 4
        if d != "Any":
            level4 = df[df["Diet"].str.lower() == d.lower()]
            if not level4.empty:
                return level4, "Matched Diet only"

        return df, "No exact match found. Showing best overall."

    # -------------------------
    # Compute TF-IDF similarity + coverage + expiry
    # -------------------------
    def compute_scores(self, df):
        matched_list, missed_list, coverage_list, expiry_list, semantic_list = [], [], [], [], []

        for idx, row in df.iterrows():
            # TF-IDF similarity (semantic approximation)
            recipe_vec = self.vectorizer.transform([row["ingredient_text"]])
            sim = cosine_similarity(recipe_vec, self.inventory_vec)[0][0]

            ingredients = [i for i in row["Cleaned_Ingredients"] if i not in IGNORE_INGREDIENTS]
            matched = [i for i in ingredients if i in self.inventory_items_filtered]
            missed = [i for i in ingredients if i not in self.inventory_items_filtered]
            coverage = len(matched) / max(len(ingredients), 1)

            # ✅ Corrected Expiry: normalized 0-1 scale
            expiry_scores = []
            for i in matched:
                days = self.inventory_dict.get(i, 30)
                normalized = max(0, (30 - days) / 30)
                expiry_scores.append(normalized)
            expiry = np.mean(expiry_scores) if expiry_scores else 0

            matched_list.append(matched)
            missed_list.append(missed)
            coverage_list.append(round(coverage,3))
            expiry_list.append(round(expiry,3))
            semantic_list.append(round(sim,3))

        df["Matched_Ingredients"] = matched_list
        df["Missed_Ingredients"] = missed_list
        df["Coverage_Score"] = coverage_list
        df["Expiry_Score"] = expiry_list
        df["Semantic_Score"] = semantic_list

        # Final Score (weighted)
        df["Final_Score"] = (
            0.4 * df["Semantic_Score"] +
            0.4 * df["Coverage_Score"] +
            0.2 * df["Expiry_Score"]
        )
        df = df.sort_values(by="Final_Score", ascending=False).reset_index(drop=True)
        return df

    # -------------------------
    # UI Widgets
    # -------------------------
    def setup_widgets(self):
        cuisines = ["Any"] + sorted(self.df_clean["Cuisine_Clean"].unique().tolist())
        diets = ["Any"] + sorted(self.df_clean["Diet"].unique().tolist())
        meals = ["Any"] + sorted(self.df_clean["MealType"].unique().tolist())

        self.cuisine_dd = widgets.Dropdown(options=cuisines, description="Cuisine:")
        self.diet_dd = widgets.Dropdown(options=diets, description="Diet:")
        self.meal_dd = widgets.Dropdown(options=meals, description="Meal:")
        self.recommend_btn = widgets.Button(description="Recommend Recipes", button_style="primary")
        self.next_btn = widgets.Button(description="Next Recipe", button_style="success")
        self.reset_btn = widgets.Button(description="Reset Preferences", button_style="warning")
        self.out = widgets.Output()

        self.recommend_btn.on_click(self.recommend_recipes)
        self.next_btn.on_click(self.show_next)
        self.reset_btn.on_click(self.reset_preferences)

        display(widgets.VBox([
            widgets.HBox([self.cuisine_dd, self.diet_dd, self.meal_dd]),
            widgets.HBox([self.recommend_btn, self.next_btn, self.reset_btn]),
            self.out
        ]))

    # -------------------------
    # Button callbacks
    # -------------------------
    def recommend_recipes(self, b):
        print("[INFO] Filtering recipes based on your preferences...")
        df, self.explanation = self.strict_filter(self.df_clean)
        self.filtered_df = self.compute_scores(df)
        self.index = 0
        self.show_next(None)
        print("[INFO] Recommendation ready!")

    def show_next(self, b):
        if self.filtered_df.empty:
            with self.out:
                clear_output()
                print("[WARNING] No recipes found for selected preferences.")
            return

        row = self.filtered_df.iloc[self.index % len(self.filtered_df)]
        self.index += 1
        with self.out:
            clear_output()
            print("======================================")
            print(f"Recipe: {row['RecipeName']}")
            print(f"Cuisine: {row['Cuisine_Clean']}")
            print(f"Meal Type: {row['MealType']}")
            print(f"Diet: {row['Diet']}")
            print(f"Coverage Score: {row['Coverage_Score']}")
            print(f"Expiry Score: {row['Expiry_Score']}")
            print(f"Semantic Score: {row['Semantic_Score']}")
            print(f"Final Score: {round(row['Final_Score'],3)}")
            print(f"Matched Ingredients: {', '.join(row['Matched_Ingredients'])}")
            print(f"Missed Ingredients: {', '.join(row['Missed_Ingredients'])}")
            print(f"Why this recipe? {self.explanation}")
            instructions = row.get("TranslatedInstructions", None)
            if instructions and pd.notna(instructions):
                print("\nInstructions:\n", instructions)

    def reset_preferences(self, b):
        self.cuisine_dd.value = "Any"
        self.diet_dd.value = "Any"
        self.meal_dd.value = "Any"
        self.index = 0
        with self.out:
            clear_output()
            print("[INFO] Preferences reset. Choose new options and click Recommend Recipes.")
```
